```python
#Todo 

class Node:
    def __init__(self, dataval=None, prevval = None, nextvalpass = None, nextvalfail = None, decisionfunc = None):
        self.dataval = dataval
        self.prevval = None
        self.nextvalpass = None
        self.nextvalfail = None
        self.decisionfunc = None
    def PrintNodeInDetail(self):
        print("Printing Node In Detail:")
        print("Node Data = ", self.dataval)
        if(self.prevval != None):
            print("Prev Node = ", self.prevval.dataval)
        else:
            print("Prev Node = \"None\"")
        if(self.nextvalpass != None):
            print("Next Node Pass = ", self.nextvalpass.dataval)
        else:
            print("Next Node Pass = \"None\" ")
        if(self.nextvalfail != None):
            print("Next Node Fail = ", self.nextvalfail.dataval)
        else:
            print("Next Node Fail = \"None\" ")
        print("Finished Printing")
    def SetPrev(self, prev_node):
        self.prevval = prev_node
    def SetNext(self, next_node):
        self.nextval = next_node
    

class DecisionTree:
    def __init__(self):
        self.rootnode = None

    # Print the linked list
    def listprintallpass(self):
        print("***Printing Tree Only Passes***")
        printval = self.rootnode
        print("ROOT NODE")
        printval.PrintNodeInDetail()
        printval = printval.nextvalpass
        while printval is not None:
            print("Passes INTO")
            printval.PrintNodeInDetail()
            printval = printval.nextvalpass
        print("***Finished Printing Tree Only Passes***")
    def listprintallfail(self):
        print("***Printing Tree Only Fails***")
        printval = self.rootnode
        print("ROOT NODE")
        printval.PrintNodeInDetail()
        printval = printval.nextvalfail
        while printval is not None:
            print("FAILS INTO")
            printval.PrintNodeInDetail()
            printval = printval.nextvalfail
        print("***Finished Printing Tree Only Fails***")
    def NewNodeToFront(self,newdata,root_becomes_pass):
        NewNode = Node(newdata)
        # Update the new nodes next val to existing node
        self.rootnode.prevval = NewNode
        if(root_becomes_pass):
            NewNode.nextvalpass = self.rootnode
            NewNode.nextvalfail = None
        else:
            NewNode.nextvalfail = self.rootnode
            NewNode.nextvalpass = None

        self.rootnode = NewNode
        
    

my_decision_tree = DecisionTree()
e1 = Node("Mon")
my_decision_tree.rootnode = e1

e2 = Node("Tue")
e3 = Node("Wed")

e4 = Node("Pancake")
e5 = Node("Waffle")


e1.nextvalfail = e2
e2.nextvalfail = e3

e1.nextvalpass = e4
e4.nextvalpass = e5


# my_decision_tree.listprintallfail()

my_decision_tree.listprintallpass()


# my_list.rootnode.nextval = e2
# e2.nextval = e3
# e2.prevval = e1

# e3.prevval =e2

# my_list.NewNodeToFront("Sat", True)

# my_list.listprintallpass()

# print(e3.prevval.dataval)

# e1.SetNext(None)
# e1.SetPrev(None)

# e1.PrintNodeInDetail()
```
