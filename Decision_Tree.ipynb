```python
#Todo 

class Node:
    def __init__(self, dataval=None, prevnode = None, nextnodepass = None, nextnodefail = None, decisionfunc = None):
        self.dataval = dataval
        self.prevnode = None
        self.nextnodepass = None
        self.nextnodefail = None
        self.decisionfunc = decisionfunc
    def PrintNodeInDetail(self):
        print("Printing Node In Detail:")
        print("Node Data = ", self.dataval)
        if(self.prevnode != None):
            print("Prev Node = ", self.prevnode.dataval)
            print("Prev Node Reason To Pass = ", self.PrevNodeReasonToPass())
        else:
            print("Prev Node = \"None\"")
        if(self.nextnodepass != None):
            print("Next Node Pass = ", self.nextnodepass.dataval)
        else:
            print("Next Node Pass = \"None\" ")
        if(self.nextnodefail != None):
            print("Next Node Fail = ", self.nextnodefail.dataval)
        else:
            print("Next Node Fail = \"None\" ")
        print("Finished Printing")
    def SetPrev(self, prev_node):
        self.prevnode = prev_node
    def SetNext(self, next_node):
        self.nextnode = next_node
    def PrevNodeReasonToPass(self):
        if(self.prevnode == None):
            return "No Previous Node"
        elif(self.prevnode.nextnodepass == self and self.prevnode.nextnodefail != self):
            return "Pass Case"
        elif(self.prevnode.nextnodepass != self and self.prevnode.nextnodefail == self):
            return "Fail Case"
        elif(self.prevnode.nextnodepass == self and self.prevnode.nextnodefail == self):
            return "Pass And Fail Case"
        else:
            print("error prev node doesn't point to current node, broken list")
            exit()
    def EvaluateDecisionFunc(self, arg):
        return self.decisionfunc(arg)

class DecisionTree:
    def __init__(self):
        self.rootnode = None

    # Print the linked list
    def listprintallpass(self):
        print("***Printing Tree Only Passes***")
        printval = self.rootnode
        print("ROOT NODE")
        printval.PrintNodeInDetail()
        printval = printval.nextnodepass
        while printval is not None:
            print("Passes INTO")
            printval.PrintNodeInDetail()
            printval = printval.nextnodepass
        print("***Finished Printing Tree Only Passes***")
    def listprintallfail(self):
        print("***Printing Tree Only Fails***")
        printval = self.rootnode
        print("ROOT NODE")
        printval.PrintNodeInDetail()
        printval = printval.nextnodefail
        while printval is not None:
            print("FAILS INTO")
            printval.PrintNodeInDetail()
            printval = printval.nextnodefail
        print("***Finished Printing Tree Only Fails***")
    def listprintcustomroute(self, path):
        print("***Printing Tree Custom Path***")
        printval = self.rootnode
        print("ROOT NODE")
        printval.PrintNodeInDetail()
        
        path_choice_cursor = 0
        choice = path[path_choice_cursor]
        
        if(choice == False):
            printval = printval.nextnodefail
        else:
            printval = printval.nextnodepass
        while printval is not None and path_choice_cursor < len(path):
            if(choice == False):   
                print("FAILS INTO")
            else:
                print("PASSES INTO")
            
            printval.PrintNodeInDetail()
            path_choice_cursor += 1
            choice = path[path_choice_cursor]
            if(choice == False):
                printval = printval.nextnodefail
            else:
                printval = printval.nextnodepass
        print("***Finished Printing Tree Custom Path***")
    def NewNodeToFront(self,newdata,root_becomes_pass):
        NewNode = Node(newdata)
        # Update the new nodes next val to existing node
        self.rootnode.prevnode = NewNode
        if(root_becomes_pass):
            NewNode.nextnodepass = self.rootnode
            NewNode.nextnodefail = None
        else:
            NewNode.nextnodefail = self.rootnode
            NewNode.nextnodepass = None

        self.rootnode = NewNode
        

def greater_than_10(number):
    
    if(type(number) != int and type(number) != float):
        print("error func \"greater_than_10\" expected type int or float got type: ", type(number))
#         exit()
    else:
        if(number > 10):
            return True
        else:
            return False
        


my_decision_tree = DecisionTree()
e1 = Node("Mon", None, None, None, greater_than_10)


print(e1.EvaluateDecisionFunc(6))



# my_decision_tree.rootnode = e1

# e2 = Node("Pizza")
# e3 = Node("Sausage")

# e4 = Node("Pancake")
# e5 = Node("Waffle")


# e1.nextnodefail = e2
# e2.nextnodepass = e3
# e3.nextnodefail = e4
# e4.nextnodefail = e5

# my_decision_tree.listprintallfail()

# my_decision_tree.listprintallpass()

my_decision_tree.listprintcustomroute([False, True, False, False, True])


# my_list.rootnode.nextnode = e2
# e2.nextnode = e3
# e2.prevnode = e1

# e3.prevnode =e2

# my_list.NewNodeToFront("Sat", True)

# my_list.listprintallpass()

# print(e3.prevnode.dataval)

# e1.SetNext(None)
# e1.SetPrev(None)

# e1.PrintNodeInDetail()
```
